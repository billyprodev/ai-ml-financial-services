````python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
data = pd.DataFrame({
    'TransactionAmount': np.random.uniform(10, 1000, 100),
    'AccountBalance': np.random.uniform(500, 5000, 100),
    'TransactionFrequency': np.random.poisson(5, 100)
})

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Choose the number of clusters (you may want to experiment with this)
num_clusters = 3

# Apply KMeans clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_data)

# Visualize the clusters (for two features)
plt.scatter(data['TransactionAmount'], data['AccountBalance'], c=data['Cluster'], cmap='viridis')
plt.title('Customer Segmentation')
plt.xlabel('Transaction Amount')
plt.ylabel('Account Balance')
plt.show()

# Display the cluster centers (in the standardized feature space)
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_centers_df = pd.DataFrame(cluster_centers, columns=data.columns[:-1])
print(cluster_centers_df)
````

````python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler,LabelEncoder
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import seaborn as sns

from sklearn.metrics import silhouette_score

# Load the dataset
file_path = '/content/drive/MyDrive/Colab Notebooks/Mall_Customers.csv'  # Replace with the actual file path
data_headers = pd.read_csv(file_path, low_memory=False, nrows=5)
listAllColumns = data_headers.columns.tolist()
print("\n")
print("Let's list all the dataset columns.")
print(listAllColumns)

##TASK: 1. Data Exploration and Analysis (20 points)
##  Background: Understanding the dataset's structure and key characteristics.
##  Tasks:
##    Explore the distribution of features like 'Annual Income' and 'Spending Score'.
##    Identify any interesting patterns or insights from the data.
##  Points: Detailed analysis and insights from data exploration.

#BR: Identify all numeric and non-numeric columns
num_cols=data_headers._get_numeric_data().columns
print("\n")
print("Let's list all the numerical columns.")
print(list(num_cols))
cat_cols=set(data_headers)-set(num_cols)
print("\n")
print("Let's list all non-numerical columns.")
print(list(cat_cols))

#BR: Import the whole dataset now
data = pd.read_csv(file_path, low_memory=False)
listColumns = data.columns.tolist()

# Print the top and bottom 5 rows
print("\n")
print("Let's examine visually the first few rows.")
print(data.head(10))
print("\n")
print("Let's examine visually the last few rows.")
print(data.tail(10))

print("\n")
print("Let's see some summary statistic about the data for numeric columns.")
print("\n")
print(data.describe())

print("\n")
print("Let's see some frequency statistic about the data for non-numeric columns.")
print("\n")
print(data['Gender'].value_counts(normalize=True))

#BR: Let's check for missing values in the dataset
missing_values = data.isnull().sum()
missing_percentage = (data.isnull().sum() / len(data)) * 100

#BR: Formatting analysis for display
missing_data_analysis = pd.DataFrame({
    'Feature': data.columns,
    'Missing Values': missing_values,
    'Percentage (%)': missing_percentage
})

print("\n")
#BR: To display the analysis results
print("Missing Data Analysis for the Entire Dataset:")
print(missing_data_analysis)
print("\n")
print("No missing values to handle by imputing or doing one-hot encoding.")


numerical_cols = num_cols
print("\n--- Numerical Variable Chart Analysis ---")
for col in numerical_cols:
    plt.figure(figsize=(10, 8))
    sns.histplot(x=col, data=data)
    plt.title(f'Distribution of {col}')
    plt.show()

print("\n--- Categorical Variable Chart Analysis ---")
# Order the histogram labels for a specific column
#ordered_labels = ['value1','value2']
#plt.figure(figsize=(10, 8))
#sns.countplot(x='column_name', data=data, order=ordered_labels)
#plt.title(f'Distribution of column_name')
#plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
#plt.show()

categorical_cols = cat_cols

for col in categorical_cols:
    plt.figure(figsize=(10, 8))
    sns.countplot(x=col, data=data)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45, ha='right')
    plt.show()


# Encoding categorical variables (Gender is the only categorical value here)
le = LabelEncoder()

print("\n")
##BR: quick script to check for non-numerical columns. The following code shows there are categorical variables.
categorical_cols = data.select_dtypes(exclude=["number","bool_"]).columns
if len(categorical_cols) > 0:
    print("Categorical columns found:")
    print(categorical_cols)
    # data = pd.get_dummies(data, columns=categorical_cols, drop_first=True)  #drop_first to avoid multicollinearity
    for col in categorical_cols:
        data[col]= le.fit_transform(data[col])
    # print("\nData after creating dummy variables:")
    print("\nData after label encoding categorical data:")
    print(data.head())
else:
    print("No categorical columns found.")


print("\n")
# Heatmap visualization
print("\n--- Heatmap of Correlation Matrix ---")
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()


print("\n")
print("Let's focus on the relevant features.")
# Select relevant features for clustering
selected_features = data[['Annual Income (k$)', 'Spending Score (1-100)']]

print("\n")
print("Time to standardize the features we will use.")
# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(selected_features)


##TASK: 2. Enhanced Clustering with K-Means (20 points)
##  Background: Application of K-Means clustering to the standardized data.
##  Tasks:
##  Experiment with different numbers of clusters.
##  Justify the chosen number of clusters based on your analysis.
##  Points: Creativity in experimenting with cluster numbers and justification.

# Apply KMeans clustering
# This can be adjusted based on experimentation
print("\n")
print("Let's apply K-Means clustering and visualize 3 separate examples.")
kmeans3 = KMeans(n_clusters=3, random_state=42)
data['Cluster3'] = kmeans3.fit_predict(scaled_data)
kmeans4 = KMeans(n_clusters=4, random_state=42)
data['Cluster4'] = kmeans4.fit_predict(scaled_data)
kmeans5 = KMeans(n_clusters=5, random_state=42)
data['Cluster5'] = kmeans5.fit_predict(scaled_data)


# Visualize the clusters (for two features, 3 clusters)
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], c=data['Cluster3'], cmap='viridis')
plt.title('Customer Segmentation (3)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()

# Visualize the clusters (for two features, 4 clusters)
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], c=data['Cluster4'], cmap='viridis')
plt.title('Customer Segmentation (4)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()

# Visualize the clusters (for two features, 5 clusters)
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], c=data['Cluster5'], cmap='viridis')
plt.title('Customer Segmentation (5)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()

print("\nLet's now build and compare a range of clusters from 1 to 11 to see the optimal number of clusters using the Elbow Method.")
# Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

print("\n The elbow is located at number of clusters =5! ")

print("\n Let's now try again to do the same but this time let's look at the Silouette Score.")
silhouette_scores = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_data)
    score = silhouette_score(scaled_data, kmeans.labels_)
    silhouette_scores.append(score)

plt.plot(range(2, 11), silhouette_scores)
plt.title('Silhouette Score for each number of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()

print("\n The silhouette score is highest at number of clusters =5 again! ")
print("\n Both method converge with the understanding we should choose a cluster of 5. ")

kmeans = KMeans(n_clusters=5, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_data)

##TASK: 3. Advanced Implementation of Hierarchical Clustering (20 points)
##  Background: Applying and interpreting Hierarchical Clustering.
##  Tasks:
##  Analyze the dendrogram to determine the optimal number of clusters.
##  Provide insights into how hierarchical clustering segments customers differently than K-Means.
##  Points: Quality of analysis and interpretation of hierarchical clustering results.

print("\n Let's now apply and interpret hierarchical clustering.")
# Prepare for Hierarchical Clustering
Z = linkage(scaled_data, 'ward')

# Dendrogram visualization
plt.figure(figsize=(10, 8))
dendrogram(Z)
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Distance')
#plt.legend(data['Cluster'])
plt.xticks(rotation=90, ha='right') # Rotate x-axis labels for better readability
plt.show()

print("The y-axis of the Dendogram is a measure of closeness of either individual data points or clusters.")
print("\nWhen we observe at distance 5 or below, we noticed about 5 major clusters.")
print("This tree form representation of clusters reaffirms what we learned from the Elbow method but gives us additional insight of potential clusters sizes.")


##TASK: 4. Comprehensive Cluster Analysis (20 points)
##      Background: Deeper evaluation of the clustering results.
##      Tasks:
##      Compare the characteristics of the clusters from both clustering methods.
##      Calculate and interpret additional cluster quality metrics beyond silhouette score.
##      Points: Depth of comparative analysis and use of additional metrics.

silhouette_avg = silhouette_score(scaled_data, kmeans.labels_)
print("\nSilhouette Score for K-Means:", silhouette_avg)

# Display the cluster centers (in the standardized feature space)
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
# Fix: Use the column names from selected_features
cluster_centers_df = pd.DataFrame(cluster_centers, columns=selected_features.columns)
print("\nCluster Centers (Original Scale):") # Changed title to reflect original scale
print(cluster_centers_df)


##TASK: 5. Insightful Reporting and Visualization (20 points)
##      Background: Effective communication of findings.
##      Tasks:
##      Create insightful visualizations to represent the clustering results.
##      Write a comprehensive report discussing findings, methodologies, and potential business implications.
##      Points: Clarity, creativity, and thoroughness in reporting and visualization.

plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], c=kmeans.labels_)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()


print("Let's analyze the characteristics of the 5 clusters identified by K-Means.")

# Group data by cluster and calculate mean of features
cluster_analysis = data.groupby('Cluster')[selected_features.columns].mean()
print("Mean values of features per cluster:")
print(cluster_analysis)

# Add the number of customers in each cluster
cluster_counts = data['Cluster'].value_counts().reset_index()
cluster_counts.columns = ['Cluster', 'Number of Customers']
cluster_analysis = pd.merge(cluster_analysis, cluster_counts, on='Cluster')
print("Cluster analysis including customer counts:")
print(cluster_analysis)

print("Interpretation of Clusters:")
for index, row in cluster_analysis.iterrows():
  print("\n")
  print(f"Cluster {int(row['Cluster'])}:")
  print(f"Number of Customers: {int(row['Number of Customers'])}")
  print(f"Average Annual Income: ${row['Annual Income (k$)']:.2f}k")
  print(f"Average Spending Score: {row['Spending Score (1-100)']:.2f}")

  # Provide descriptive labels based on the mean values
  if row['Annual Income (k$)'] > 70 and row['Spending Score (1-100)'] > 70:
    print("Description: High Income, High Spending (Potential 'Target' Customers)")
  elif row['Annual Income (k$)'] > 70 and row['Spending Score (1-100)'] < 40:
    print("Description: High Income, Low Spending (Potential 'Careful' Customers)")
  elif row['Annual Income (k$)'] < 40 and row['Spending Score (1-100)'] > 70:
     print("Description: Low Income, High Spending (Potential 'Impulsive' Customers)")
  elif row['Annual Income (k$)'] < 40 and row['Spending Score (1-100)'] < 40:
     print("Description: Low Income, Low Spending (Potential 'Frugal' Customers)")
  else:
    print("Description: Mid-Range Income and Spending (Potential 'Average' Customers)")

print("We can now take each of these groups and design a specific campaign to attract each category of buyer to the product or service.")
````

### Introduction to Cluster Evaluation Techniques

In the realm of unsupervised machine learning, determining the optimal number of clusters is a pivotal decision that can significantly impact the outcomes of your model. Cluster evaluation techniques are essential tools that provide guidance in this decision-making process. Two of the most widely recognized methods for evaluating clustering results are the Elbow Method and the Silhouette Score.

#### Elbow Method
- **Explanation**: The Elbow Method is a heuristic used in determining the number of clusters in a data set. The approach involves plotting the explained variance as a function of the number of clusters, and picking the point where the increase in variance explained by adding another cluster is not significant anymore. This point is known as the 'elbow', where the graph bends.
- **Interpretation**: In the Elbow Method, one should look for a change in the gradient of the line plot; a sharp change like an elbow suggests the optimal number of clusters. The idea is that adding more clusters beyond this number does not provide much better modeling of the data.

#### Silhouette Score
- **Explanation**: The Silhouette Score is a metric used to calculate the goodness of a clustering technique. It measures the distance between points within a cluster and the distance to points in the next nearest cluster. The score ranges from -1 to +1, where a high value indicates that the points are well clustered.
- **Interpretation**: A Silhouette Score close to +1 indicates that the sample is far away from the neighboring clusters. A value of 0 indicates that the sample is on or very close to the decision boundary between two neighboring clusters and negative values indicate that those samples might have been assigned to the wrong cluster.

Both methods provide different lenses through which to view the clustering results and can be used in conjunction to make a more informed decision. The Elbow Method gives us an insight into the variance within each cluster, whereas the Silhouette Score provides a measure of how similar an object is to its own cluster compared to others. The optimal number of clusters is often the one that balances between the two measures, subject to the specific context and use case of the analysis.

````python
##Task: Advanced Clustering Techniques (20 points)
##      DBSCAN Clustering:
##      Implement DBSCAN and compare its segmentation with K-Means and Hierarchical clustering.
##      Analyze the clusters formed by DBSCAN for any unique characteristics.
##      Points: Quality of implementation and comparative analysis.


import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA

# Apply DBSCAN clustering
# A starting point could be to look at the distance to the k-th nearest neighbor plot for determining epsilon.
dbscan = DBSCAN(eps=0.5, min_samples=5)
data['DBSCAN_Cluster'] = dbscan.fit_predict(scaled_data)

# Visualize the clusters formed by DBSCAN
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], c=data['DBSCAN_Cluster'], cmap='viridis')
plt.title('Customer Segmentation (DBSCAN)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()

# Analyze the clusters formed by DBSCAN
# DBSCAN assigns points it considers noise to cluster -1.
print("DBSCAN Cluster Analysis:")
print(data['DBSCAN_Cluster'].value_counts())

# Compare DBSCAN segmentation with K-Means and Hierarchical Clustering
print("Comparison of Clustering Methods:")
print("K-Means clusters (5 clusters):")
print(data['Cluster'].value_counts().sort_index())
# Hierarchical clustering labels would need to be added to the dataframe after cutting the dendrogram.
# For comparison, we can visually inspect the dendrogram against the DBSCAN clusters.

# Analyze unique characteristics of DBSCAN clusters
# Filter out noise points (-1) for analysis
dbscan_clusters_df = data[data['DBSCAN_Cluster'] != -1]

dbscan_cluster_analysis = dbscan_clusters_df.groupby('DBSCAN_Cluster')[selected_features.columns].mean()
print("Mean values of features per DBSCAN cluster (excluding noise):")
print(dbscan_cluster_analysis)


print("Unique characteristics of DBSCAN compared to K-Means/Hierarchical:")
print("- DBSCAN can identify clusters of arbitrary shape, not just spherical like K-Means.")
print("- DBSCAN does not require specifying the number of clusters beforehand.")
print("- DBSCAN identifies noise points (assigned to cluster -1).")
print("- The results of DBSCAN are highly sensitive to the parameters `eps` and `min_samples`.")
print("- `eps` (epsilon): The maximum distance between two samples for one to be considered as in the neighborhood of the other.")
print("- `min_samples`: The number of samples in a neighborhood for a point to be considered as a core point.")
print("- Unlike K-Means which forces every point into a cluster, DBSCAN can leave some points unassigned (noise).")

# Further analysis could involve looking at the spread (standard deviation) within each DBSCAN cluster
# to see how dense they are.

dbscan_cluster_std = dbscan_clusters_df.groupby('DBSCAN_Cluster')[selected_features.columns].std()
print("Standard deviation of features per DBSCAN cluster (indicating density/spread):")
print(dbscan_cluster_std)

print("\n")
##Task: Dimensionality Reduction Technique (15 points)
##      Principal Component Analysis (PCA):
##      Apply PCA to the data and visualize the results.
##      Discuss how dimensionality reduction impacts the clustering results and its potential use in simplifying complex datasets.
##      Points: Effectiveness of PCA implementation and analysis of its impact.

# Apply PCA to the scaled data
# Choose the number of components (e.g., 2 for visualization)
n_components = 2
pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(scaled_data)

# Create a DataFrame with the principal components
pca_df = pd.DataFrame(data=principal_components, columns=[f'principal_component_{i+1}' for i in range(n_components)])

# Add the cluster labels from K-Means to the PCA DataFrame for visualization
pca_df['Cluster'] = data['Cluster']

# Visualize the results of PCA with K-Means clusters
plt.figure(figsize=(10, 8))
plt.scatter(pca_df['principal_component_1'], pca_df['principal_component_2'], c=pca_df['Cluster'], cmap='viridis')
plt.title('PCA of Customer Data with K-Means Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Explained variance ratio by each principal component
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio by Principal Components:")
for i, ratio in enumerate(explained_variance_ratio):
    print(f"Principal Component {i+1}: {ratio:.4f}")

print("Total Explained Variance by the selected components:", explained_variance_ratio.sum())
print("\n")
print("Impact of Dimensionality Reduction (PCA) on Clustering and its Usefulness:")
print("1. Simplification and Visualization:")
print("- PCA reduces the number of features while retaining most of the variance.")
print("- For datasets with many features, reducing to 2 or 3 components allows for easy visualization of the data and clusters in a 2D or 3D plot.")
print("- As seen in the plot above, the clusters are visualized in the reduced 2D space.")
print("2. Noise Reduction:")
print("- PCA can help in reducing noise in the data by discarding components with low variance, which often correspond to noise.")
print("- This can potentially lead to better and more robust clustering results by focusing on the most important underlying patterns.")
print("3. Computational Efficiency:")
print("- Clustering algorithms can be computationally expensive, especially with a high number of dimensions.")
print("- Applying clustering on a lower-dimensional dataset (after PCA) can significantly reduce computation time and memory usage.")
print("4. Lowering the number of dimensions to 2 to make the data easier to interpret:")
print("- In high-dimensional spaces, data points become increasingly sparse, making distance-based clustering algorithms (like K-Means) less effective.")
print("- PCA can help mitigate this issue by projecting the data into a lower-dimensional space where points are closer to each other.")
print("5. Potential Loss of Information:")
print("- While PCA aims to retain most variance, some information is inevitably lost when dimensions are reduced.")
print("- The choice of the number of components is crucial; selecting too few components might discard important variations relevant to clustering.")
print(" The visualization in 2D (using 2 components) might not perfectly represent the original clusters if the remaining variance is significant.")
print("\n")
print("Overall, PCA can be a valuable preprocessing step for clustering, especially for high-dimensional data. It aids in visualization, reduces noise and computation,")
print("and can help improve clustering performance by addressing the curse of dimensionality. However, the choice of the number of components needs")
print("careful consideration to avoid losing too much important information.")
````
